```python
def make2():
    if board[4] not in X and board[4] not in O:
        return 4
    elif board[1] not in X and board[1] not in O:
        return 1
    elif board[3] not in X and board[3] not in O:
        return 3    
    elif board[5] not in X and board[5] not in O:
        return 5
    else:
        return 7
```

```python
def posswin(p):
    if p == 'X':
        for x1 in X:
            for x2 in X:
                if (15-(x1+x2)) > 0 and (15-(x1+x2)) <= 9:
                    if (15-(x1+x2)) not in X and (15-(x1+x2)) not in O:
                        for i in xrange(0,9):
                            if board[i] == (15-(x1+x2)):
                                return i
        return 0
    
    if p == 'O':
        for o1 in O:
            for o2 in O:
                if (15-(o1+o2)) > 0 and (15-(o1+o2)) <= 9:                    
                    if (15-(o1+o2)) not in X and (15-(o1+o2)) not in O:
                        for i in xrange(0,9):
                            if board[i] == (15-(o1+o2)):
                                return i
        return 0
    
    
```

```python
def go(n,turn):
    if turn[0] % 2 == 1:
        X.append(board[n])
    else:
        O.append(board[n])
    turn[0] = turn[0] + 1
```
